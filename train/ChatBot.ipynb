```python
from transformers import AutoTokenizer
import torch
import pandas as pd
from sklearn.model_selection import train_test_split

# 데이터 준비
data = [
    {"text": "2023년 매출 데이터를 보여줘.", "label": 0},
    {"text": "A 회사의 성과는 어떠한가요?", "label": 1},
    {"text": "B 회사의 출고 상태를 알려줘.", "label": 0},
    {"text": "출고 상태는 무엇을 의미하나요?", "label": 1},
    {"text": "2022년 매출을 요약해줘.", "label": 0},
    {"text": "성과 평가의 기준은 무엇인가요?", "label": 1},
]

data = pd.read_csv('/drive/MyDrive/BigProject/')

# 데이터프레임 생성
df = pd.DataFrame(data)

# 학습/검증 데이터 분리
train_texts, val_texts, train_labels, val_labels = train_test_split(
    df["text"], df["label"], test_size=0.2, random_state=42
)

# DistilBERT 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

# 토큰화 함수
def tokenize_function(texts, labels):
    tokens = tokenizer(
        texts.tolist(),  # Pandas Series -> 리스트 변환
        padding="max_length",
        truncation=True,
        max_length=128,
        return_tensors="pt"
    )
    tokens["labels"] = torch.tensor(labels.tolist())  # 리스트로 처리
    return tokens

# 학습/검증 데이터 토큰화
train_tokens = tokenize_function(train_texts, train_labels)
val_tokens = tokenize_function(val_texts, val_labels)
```

```python
from transformers import AutoModelForSequenceClassification

# DistilBERT 분류 모델 로드
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)
```

```python
from transformers import Trainer, TrainingArguments
import torch

# GPU 설정
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# 학습 설정
training_args = TrainingArguments(
    output_dir="./results",         # 모델 저장 경로
    evaluation_strategy="epoch",   # 검증 주기
    logging_dir="./logs",           # 로그 저장 경로
    per_device_train_batch_size=8, # 배치 크기
    per_device_eval_batch_size=8,
    num_train_epochs=5,            # 학습 에포크 수
    learning_rate=5e-5,            # 학습률
    save_strategy="epoch",         # 저장 주기
    logging_steps=10,              # 로그 출력 주기
    report_to="none"  # W&B 비활성화
)
```

```python
from torch.utils.data import Dataset

# PyTorch Dataset 정의
class CustomDataset(Dataset):
    def __init__(self, tokens):
        self.tokens = tokens

    def __len__(self):
        return len(self.tokens["input_ids"])

    def __getitem__(self, idx):
        return {
            "input_ids": self.tokens["input_ids"][idx],
            "attention_mask": self.tokens["attention_mask"][idx],
            "labels": self.tokens["labels"][idx]
        }

# 학습/검증 데이터셋 생성
train_dataset = CustomDataset(train_tokens)
val_dataset = CustomDataset(val_tokens)

# Trainer 객체 생성
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset
)
```

```python
# 모델 학습
trainer.train()
```

```python
# 모델 저장
model.save_pretrained("./text_to_sql_classifier")
tokenizer.save_pretrained("./text_to_sql_classifier")
```

```python
# BERT 땡겨오기.
import torch
# Load model directly
from transformers import AutoTokenizer, AutoModelForMaskedLM, AutoModelForSequenceClassification

# DistilBERT 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# 토큰화 함수
def tokenize_function(texts, labels):
    tokens = tokenizer(
        texts.tolist(),  # Pandas Series -> 리스트 변환
        padding="max_length",
        truncation=True,
        max_length=128,
        return_tensors="pt"

    )
    tokens["labels"] = torch.tensor(labels.tolist())  # 리스트로 처리
    return tokens

# 학습/검증 데이터 토큰화
train_tokens = tokenize_function(train_texts, train_labels)
val_tokens = tokenize_function(val_texts, val_labels)

# DistilBERT 분류 모델 로드
model = AutoModelForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)
```
